**Objekt**

De flesta moderna programmeringsspråk är objektiorienterande. Med det menas att allt som hanteras i språket är objekt. Ett objekt är en datastruktur med tillhörande tillståndsvariabler och metoder. Vilka tillståndsvariabler och metoder som hör till objektet beskrivs i klassen som objektet är en instans av. Varje objekt har dock olika värden på sina dataattribut. Objektorienterad programmering handlar om att bryta ner uppgifter till klasser, och objekt. Objekt i programmering är ofta kopplade till något verkligt objekt. Ex: kunder, bilar m.m.

**Klass**

En klass beskriver mallen/ritningen till ett objekt. Ett objekt är en instans av en klass. Ett objekt har instansvariabler. Instansvariabler anger egenskaper hos objektet. Att skapa ett objekt av en klass kallas att instansiera objektet.

En klass kan innehålla:

- Dokumentation (doc-string)
- Tillståndsvariabler
- Konstanter (Skrivs med versaler ex PI)
- metoder
    - __init__ (kontruktor-metoden skapar objektet)
    - __str__ (str-metoden skriver ut information om objektet)
    - egendefinierade metoder
    - hämta och sätta metoder (get/set-metoder används för att ändra och hämta ett objekts tillståndsvariabler)

**Exempel**

        # Objektet 'Kalle Anka' är en instans av String-klassen 
        # och har variabeln string1 (string1 pekar på minnesplatsen)
        string1 = 'Kalle Anka'

        #Funktionen len körs på objektet med namn string1
        len(string1)

        #Metoden lower som tillhör klassen string körs på objektet med namn string1
        string1.lower()

Funktioner är inte kopplade till ett objekt. Metoder är alltid kopplade till ett objekt. Fråga är nedanstående en metod eller en funktion 

    Random.randint()

Klass-namn börjar alltid med versal. I python skrivs egan klasser genom att använda

    class Min_klass:
    """Doc-sträng för klassen"""
    
    KONST1 = 12
    
    def __init__(self, parameter1, parameter2):
        """Doc-sträng för konstrukorn."""
        Konstruktor-metoden instansierar ett objektet av klassen"""
        self.parameter1 = parameter1
        self.parameter2 = parameter2

        """Doc-sträng för metoden."""
