```python
#Tidigare

import sys
from PyQt4.QtCore import *
from PyQt4.QtGui import *
import copy

app = QApplication(sys.argv)

#QMainWindow ger oss tillgång till andra funktioner som menyer, statusrad mm
class MySort(QMainWindow):
    
    def __init__(self, parent=None):
        super(MySort, self).__init__()
        
        self.setWindowTitle("My Sort") 
        self.initUI()
        
        self.lista = []
    
    def initUI(self):
        """Initierar alla komponenter"""
        
        self.frame = QWidget(self)
        self.setCentralWidget(self.frame)
        
        #Hela fönstret
        self.layout = QVBoxLayout()
        
        #Widgets
        self.input_led = QLineEdit()
        self.layout.addWidget(self.input_led)
        self.input_btn = QPushButton("Lägg till")
        self.input_btn.clicked.connect(self.btn_input_action)        
        
        self.layout.addWidget(self.input_btn)
        self.output_led = QLineEdit()
        self.layout.addWidget(self.output_led)
        
        #Buttons
        self.box1 = QHBoxLayout()
    
        self.bubble_text = QTextEdit()
        self.box1.addWidget(self.bubble_text)        
        
        self.insert_text = QTextEdit()
        self.box1.addWidget(self.insert_text)        
        self.layout.addLayout(self.box1)        
        
        self.box2 = QHBoxLayout()        
        
        self.bubble_btn = QPushButton('Bubble', self)
        self.bubble_btn.clicked.connect(self.btn_bubble_action)
        self.box2.addWidget(self.bubble_btn)
        
        self.insert_btn = QPushButton('Insert', self)
        self.insert_btn.clicked.connect(self.btn_insert_action)
 
        self.box2.addWidget( self.insert_btn )
        self.layout.addLayout(self.box2)
        
        #Add layouts to widget
        self.frame.setLayout(self.layout)
        
       
    
    @pyqtSlot()
    def btn_input_action(self):
         print("Input")
         self.lista.append(int( self.input_led.text() ))
         string = ""
         for n in self.lista:
             string += str(n)
             string += ', '
             
         self.output_led.setText(string)
         
         self.input_led.setText("")
     
    @pyqtSlot()
    def btn_bubble_action(self):
         print("Bubble")
         alist = copy.copy(self.lista)
         for passnum in range(len(alist)-1,0,-1):
             for i in range(passnum):
                 if alist[i]>alist[i+1]:
                     temp = alist[i]
                     alist[i] = alist[i+1]
                     alist[i+1] = temp
             self.bubble_text.append(', '.join(map(str, alist)) )
    
    @pyqtSlot()
    def btn_insert_action(self):
        print("Insert")

    def run(self):
        self.show()
        sys.exit(app.exec_())

#Skapar en instans och kör den
MySort().run()
```

Datastrukturer används för att lagra information. Olika Datastrukturer används beroende på situation. Några vanliga datastrukturer är:

- listor (enkel, dubbellänkade)
- dicationaries
- Köer

**Vanliga metoder**

- .isEmpty()
- .pop()
- .put()
- .peek()

Search − Algorithm to search an item in a datastructure.

Sort − Algorithm to sort items in certain order

Insert − Algorithm to insert item in a datastructure

Update − Algorithm to update an existing item in a data structure

Delete − Algorithm to delete an existing item from a data structure

**Principer för att lägga till/ta bort element**

- FIFO First In First Out
- FILO First In Last Out

```python
#Exempel lista

class Node:
    
    def __init__(self, data):
        self.data = data
        self.nextNode = None

class List:
    
    def __init__(self):
        self.primeNode = None
        self.lastNode = None
    
    def put(self,data):
        if self.primeNode is None:
            self.primeNode = Node(data)
            return
        node = self.primeNode
        while node.nextNode is not None:
            node = node.nextNode
        node.nextNode = Node(data)
        self.lastNode = node.nextNode
        
    def peek(self):
        return self.lastNode.data
    
    def isEmpty(self):
        if self.primeNode is None:
            return True
        return False
        
def testingList():
    lista = List()
    
    #test init put
    lista.put(3)
    if lista.primeNode.data != 3:
        print("failed put")
        
    #test put
    lista.put(4)
    if lista.lastNode.data != 4:
        print("failed put")
    print("ok")
    
testingList()
```
