```python
class MyModel:
    def __init__(self):
        self.state = "ON"
    
    def updateModel(self, state):
        if state == 0:
            self.state = "OFF"
        elif state == 1:
            self.state = "ON"
    
    def getViewInfo(self):
        return self.state
    

from PyQt4.QtCore import *
from PyQt4.QtGui import *
import sys

class GUIWindow(QMainWindow):
    def __init__(self):
        QMainWindow.__init__(self)
        self.model = MyModel()
        self.initGui()
        
    def initGui(self):
        self.setWindowTitle("Gui View")
        
        frame = QWidget(self)
        self.setCentralWidget(frame)
        
        layout = QVBoxLayout()
        
        btn_on = QPushButton("State: ON")
        btn_on.clicked.connect(self.btn_on_action)
        layout.addWidget(btn_on)
        
        btn_off = QPushButton("State: OFF")
        btn_off.clicked.connect(self.btn_off_action)
        layout.addWidget(btn_off)
        
        self.led = QLineEdit()
        layout.addWidget(self.led)
        self.led.setText( self.model.getViewInfo() )
    
        frame.setLayout(layout)
        
        self.show()
        
    def btn_on_action(self):
        self.model.updateModel(1)
        self.led.setText( self.model.getViewInfo() )
    
    def btn_off_action(self):
        self.model.updateModel(0)
        self.led.setText( self.model.getViewInfo() )
        
    
class ViewCMD:
    def __init__(self):
        self.model = MyModel()
    
    def run(self):
        intp = int(input("Input: "))
        
        while intp is not "quit":
            self.model.updateModel(intp)
            print(self.model.getViewInfo())
            intp = int(input("Input: "))

def runCMD():
    ViewCMD().run()
    
def runGUI():
    app = QApplication(sys.argv)
    main = GUIWindow()
    sys.exit(app.exec_())
    
#runGUI()
runCMD()
```
